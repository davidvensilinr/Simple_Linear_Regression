**Simple Linear Regression Model for predicting the number of hours needed to study for getting particular score**

Let's import the necessary packages such as

* Pandas- Used for handling the dataset, cleaning and pre processing
* Numpy - Used for converting the data from pandas to arrays and perform mathematical and other functions
* Matplotlib - For developing a visual representation of our data to get better understanding of our data
* linear_model - Linear Regression model



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

Let's convert our data in csv file into DataFrame using Pandas

```python
df=pd.read_csv('scores.csv')
```

```python
df
```

Visulaizing the data by plotting in graph

```python
plt.scatter(df['Marks'],df['Time (minutes)'],color="red",marker='+')
plt.xlabel('Marks')
plt.ylabel('Time')
plt.show()
```

Let's start training our model

Basic Concept:

y=mx+c

y is the predicted value and x is our input value

m - is the slope and c is the intercept

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Marks']],df['Time (minutes)'])
```

Now since our model is trained let's compared it with our original data using matplotlib

```python
plt.scatter(df['Marks'],df['Time (minutes)'],color="red",)
plt.scatter(df['Marks'],reg.predict(df[['Marks']]),color="blue")
plt.xlabel('Marks')
plt.ylabel('Time')
plt.show()
```

```python
mark=40
pred=reg.predict([[mark]])
print(f"To get {mark}, you should be studying for {round(int(pred),0)} minutes")
```
